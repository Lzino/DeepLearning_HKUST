```python
import tensorflow as tf
```

```python
# X and Y data
x_train = [1,2,3]
y_train = [1,2,3]

W = tf.Variable(tf.random_normal([1]), name = 'weight') 
b = tf.Variable(tf.random_normal([1]), name = 'bias')

#H(x) = W*x + b
hypothesis = x_train*W + b
```

```python
#cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
#Launch the graph in a session
sess = tf.Session()

#Initializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

```python
#fit the line
for step in range(2011):
    sess.run(train)
    if step % 30 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
#Placeholder example
X = tf.placeholder(tf.float32, shape = [None])
Y = tf.placeholder(tf.float32, shape = [None])

hypothesis = X * W +b
cost2 = tf.reduce_mean(tf.square(hypothesis -Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost2)

#Launch the graph in a session
sess = tf.Session()

#Initializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2021):
    cost_val, W_val, b_val, _= sess.run([cost2,W,b,train], 
            feed_dict={X:[1,2,3,4,5], Y: [2.1,3.1,4.1,5.1,6.1]})
    if step % 30 == 0:
        print(step, cost_val, W_val, b_val)
#다른 훈련을 먼저 실행하면 이미 훈련되어서 값이 더 낮아질뿐 크게 변화는 없다.
```

```python
# Testing Model
print(sess.run(hypothesis, feed_dict = {X: [5]}))
print(sess.run(hypothesis, feed_dict = {X: [2.5]}))
print(sess.run(hypothesis, feed_dict = {X:[1.5, 3.5]}))
```
